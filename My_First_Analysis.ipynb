```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('3.2 cost_revenue_clean.csv.csv')
type(data)
```

```python
data.describe()
```

```python
5.034e+03
```

```python
X = pd.DataFrame(data,columns=['production_budget_usd']) # FEATURE OR INDEPENDENT VARIABLE
y = pd.DataFrame(data,columns=['worldwide_gross_usd']) # TARGET OR DEPENDENT VARIABLE
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y,alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.ylim(0, 3e+9)
plt.xlim(0, 4.5e+8)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X ,y)
```

Slope or theta_1

```python
regression.coef_
```

Intercept or theta_0

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y,alpha=0.3)
plt.plot(X ,regression.predict(X),color = 'red',linewidth = 4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.ylim(0, 3e+9)
plt.xlim(0, 4.5e+8)
plt.show()
```

```python
regression.score(X ,y)
```
