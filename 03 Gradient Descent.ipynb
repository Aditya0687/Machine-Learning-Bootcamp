## Gradient Descent

```python
new_x = 3
prev_x = 0
gamma = 0.1
precesion = 0.00000000001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(300):
    prev_x = new_x
    gradient = df(prev_x)
    if(gradient!=0):
        new_x = prev_x - gamma*gradient
    else:
        break
    x_list.append(new_x)
    slope_list.append(df(new_x))
    step_size = abs(new_x - prev_x)
    if(step_size<precesion):
        print("Loop ran ",n," times")
        break
print('Local Minima occurs at: ',new_x)
print('Slope at local minima: ',df(new_x))
print('Minimum cost: ',f(new_x))
```

```python
# Superimpose plot with Gradient Descent
plt.figure(figsize = (20,5))

# Chart 1: Cost Function
plt.subplot(1 ,3 ,1)

plt.title('Cost Function',fontsize = 16)
plt.xlabel('X',fontsize = 16)
plt.ylabel('f(x)',fontsize = 16)
plt.xlim(-3, 3)
plt.ylim(0,8)

plt.plot(x_1,f(x_1),color = 'b',linewidth = 4)

values = np.array(x_list)
plt.scatter(x_list,f(values),color='r',s=100,alpha =0.6)

# Chart 2: Derivative of Cost Function
plt.subplot(1 ,3 ,2)

plt.title('Derivative of Cost Function',fontsize = 16)
plt.xlabel('X',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.xlim(-2, 3)
plt.ylim(-3,6)
plt.grid()

plt.plot(x_1,df(x_1),color = 'skyblue',linewidth = 4)
plt.scatter(x_list,slope_list,color='r',s=100,alpha =0.6)

# Chart 3: Derivative of Cost Function
plt.subplot(1 ,3 ,3)

plt.title('Gradient Descent(CLOSE UP)',fontsize = 16)
plt.xlabel('X',fontsize = 16)
plt.xlim(-0.6, -0.2)
plt.ylim(-0.2,0.8)
plt.grid()

plt.plot(x_1,df(x_1),color = 'skyblue',linewidth = 4)
plt.scatter(x_list,slope_list,color='r',s=100,alpha =0.6)
plt.show()
```

## Gradient Descent as Python Function

```python
def gradient_descent(derivative_function,intial_assumption,precesion = 0.001,gamma=0.02,max_iteration = 300):
    new_x = intial_assumption
    x_list = [new_x]
    slope_list = [derivative_function(new_x)]
    for n in range(max_iteration):
        prev_x = new_x
        gradient = derivative_function(prev_x)
        new_x = prev_x - gradient*gamma
        x_list.append(new_x)
        slope_list.append(derivative_function(new_x))
        step_size = abs(new_x - prev_x)
        if(step_size<precesion):
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_function = dg, intial_assumption = 0.1,
                                                 precesion = 0.001,gamma = 0.02)
print('Local Minima occurs at: ',local_min)
print('No of steps: ',len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_function = dg, intial_assumption = -0.1)
print('Local Minima occurs at: ',local_min)
print('No of steps: ',len(list_x))
```

# Data Viz with - 3D Charts and Plots
## $$ f(x,y) = \frac{1}{3^{-x^2-y^2}+1}$$
Minimise $$ f(x,y) = \frac{1}{r + 1} $$ where r is $3^{-x^2-y^2}$

```python
def f(x,y):
    r = 3**(-x**2-y**2)
    return 1/(r+1)
```

```python
# Make data
x_4 = np.linspace(start = -2, stop = 2, num = 200)
y_4 = np.linspace(start = -2,stop = 2, num = 200)
print('Shape of X ',x_4.shape)
x_4 , y_4 = np.meshgrid(x_4 , y_4)
print('Shape of X ',x_4.shape)
```

```python
# Generating 3D Plot
fig = plt.figure(figsize=(16,12))
ax = fig.gca(projection = '3d')

ax.set_xlabel('X', fontsize =20)
ax.set_ylabel('Y', fontsize =20)
ax.set_zlabel('f(x,y) - Cost', fontsize =20)


ax.plot_surface(x_4,y_4,f(x_4,y_4), cmap = cm.coolwarm,alpha = 0.4)

plt.show()
```

# Partial Derivative and Sybolic Python Library

```python
a, b = symbols('x , y')
print('Cost function is: ', f(a,b))
print('Partial Derivative of cost function wrt x: ',diff(f(a,b), a))
print('Cost at x = 1.8 and y = 1.0 is: ', 
      f(a,b).evalf(subs={a:1.8 , b:1.0}))
print('Slope at x = 1.8 and y = 1.0 is: ',
      diff(f(a,b),a).evalf(subs={a:1.8 , b:1.0}))
```
