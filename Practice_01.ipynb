```python
# Practice of creating 3D chart
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
def f(x,y):
    r = 3**(-x**2-y**2)
    return 1/(r+1)
```

```python
# Make data
x = np.linspace(start = -2, stop = 2,num = 300)
y = np.linspace(start= -2, stop = 2,num = 300)
# z = f(x, y)
mesh_x , mesh_y = np.meshgrid(x,y)
```

```python
fig = plt.figure(figsize=[15,12])
ax = fig.gca(projection = '3d')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y) - Cost Function')

ax.plot_surface(mesh_x,mesh_y,f(mesh_x,mesh_y),cmap = cm.coolwarm,alpha = 0.5)

plt.show()
```
